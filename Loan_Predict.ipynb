```python
df = df.drop('LoanAmount', axis =1)
df.head()
```

```python
#Encoding techniques: Label Encoding and One Hot Encoding
from sklearn.preprocessing import LabelEncoder
cols = ['Gender','MartialStatus','LoanStatus']
le = LabelEncoder()
for col in cols:
  df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.dtypes
```

```python
##Split Independent and Dependent Features where loan_status is only dependent variable
X = df.drop(columns = ['LoanStatus'], axis= 1)
y = df['LoanStatus']
```

```python
df.head()
```

```python
  X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Split data into Train and Test datasheet
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 42 )
```

```python
#Logistic Regression
model1 = LogisticRegression()
model1.fit(X_train, y_train)
y_pred1 = model1.predict(X_test)
accuracy = accuracy_score(y_test, y_pred1)
```

```python
accuracy*100
```

```python
##Accuracy : the ratio of correctly predicted values to the total values
```

```python
#Cross validation on 5 sets of data
score = cross_val_score(model1, X,y, cv=5)
score
```

```python
np.mean(score)
```

```python
##DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)
accuracy = accuracy_score(y_test, y_pred2)
accuracy*100
```

```python
score = cross_val_score(model2, X,y, cv=5)
print(score)
```

```python
np.mean(score)
```

```python
##RandomClassfier
model3 = RandomForestClassifier()
model3.fit(X_train, y_train)
y_pred3 = model3.predict(X_test)
accuracy = accuracy_score(y_test, y_pred3)
accuracy*100
```

```python
##KNeibhbourClassfier
model4 = KNeighborsClassifier(n_neighbors=3)
model4.fit(X_train, y_train)
y_pred4 = model4.predict(X_test)
accuracy = accuracy_score(y_test, y_pred4)
accuracy*100
```

```python
score = cross_val_score(model4, X,y, cv=5)
print(score)
```

```python
np.mean(score*100)
```

```python
#Checking and reporing
from sklearn.metrics import classification_report

def generate_classification_report(model_name, y_test, y_pred):
  report = classification_report(y_test, y_pred)
  print(report)

generate_classification_report(model1,y_test, y_pred1)
generate_classification_report(model2,y_test, y_pred2)
generate_classification_report(model3,y_test, y_pred3)
generate_classification_report(model4,y_test, y_pred4)
```
